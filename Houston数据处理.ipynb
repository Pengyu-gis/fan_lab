```python
#把要分析的那一列数据提取出来
Tmin = pd.DataFrame()
Tmin = data['Tmin_C']

#在这列数据旁边增加一列序号
Tmin = Tmin.reset_index()
Tmin['ID'] = range(len(Tmin))

#用sorted函数排序
Tmin_sorted = Tmin.sort_values(by="Tmin_C",ascending=True) #先后按列索引1,2排序，将其 reverse 参数值改为 True，可实现降序排序
Tmin_sorted95=Tmin_sorted.head(a)
print(Tmin_sorted95)
```

```python
#把要分析的那一列数据提取出来
Tmax = pd.DataFrame()
Tmax = data['Tmax_C']

#在这列数据旁边增加一列序号
Tmax = Tmax.reset_index()
Tmax['ID'] = range(len(Tmax))

#用sorted函数排序
Tmax_sorted = Tmax.sort_values(by="Tmax_C",ascending=False) #先后按列索引1,2排序，将其 reverse 参数值改为 True，可实现降序排序
Tmax_sorted95=Tmax_sorted.head(a)
print(Tmax_sorted95)
```

```python
#高温
result1 = data[(data['Tmax_C'] >= min(Tmax_sorted95['Tmax_C'])) & (data['Date'].str.contains('2019'))]
result1.insert(loc=2,column='type',value='Tmax')
#低温
result2 = data[(data['Tmin_C'] <= max(Tmin_sorted95['Tmin_C'])) & (data['Date'].str.contains('2019'))]
result2.insert(loc=2,column='type',value='Tmin')
```

```python
data2 = pd.concat([result1,result2])
data2
```

```python
#求对应的mobility数据
path=r"G:\code\jupyterlab\houston mobility"     #批量表格所在文件路径
file=glob.glob(os.path.join(path, "daily_Houston_2019_**.csv"))      #每一个表格相同名称部分
print(file)
```

```python
mobility = pd.DataFrame()
for f in file:
    mdata=pd.read_csv(f)      #读取每个表格 #,index_col=None,encoding='ANSI',delimiter="\t",low_memory=False
    mobility = pd.concat([mobility,mdata])
    print(f)
```

```python
mobility
```

```python
mobility['Date'] = pd.to_datetime(mobility['date'],format= '%Y-%m-%d')
```

```python
mobility1 = mobility.drop(columns=['date','Unnamed: 0','visitor_flows'])
```

```python
#计算各个CTs每日的人口通量
mobility2 = mobility1.groupby(['geoid_o','Date']).sum()
mobility2
```
